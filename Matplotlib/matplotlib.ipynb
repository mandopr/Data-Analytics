## *PLOTTING FOR PYTHON DEVELOPERS*

```python
plt.style.use("fivethirtyeight")
# plot() will plot data in line representation, passed via arguments

'''
    ##### PLOT SYNTEX #####
    -> plot(x-axis-dataset, y-axis-dataset, color(linecolor), linestyle, marker, markerwidth, label(legeds label))

    #  NOTE : here some arguments which on calling function takes value don't have any specific order(label, color, linestyle etc ...)
'''

# plot for python devs
plt.plot(ages_x, py_dev_y, label="Python Devs", color="b", linewidth=3)
# plot for JS devs
plt.plot(ages_x, js_dev_y, label="JS Devs", color="y", linewidth=3)
# plot for all devs
plt.plot(ages_x, dev_y, label="All Devs", color="k", linestyle="--", marker=".", markersize=5)

# xlabel() will make a label on x-axis of plot
plt.xlabel("Age")
# ylabel() will make a label on y-axis of plot
plt.ylabel("Median Salary (USD)")
# title() will make a title for plot
plt.title("Median Salary (USD) to Age")
# legend() will return legend for specified labels passed on plot()
plt.legend()
# tight_layout() will give some padding to plot for better presentation
plt.tight_layout()
# show() will show us the plot
plt.show()
```

### SOME REAL DATA

```python
# reading a CSV file
data = pd.read_csv('../data/data.csv')
```

<br><br>
### HORIZONTAL BAR CHART

```python
plt.style.use("fivethirtyeight")
# selecting ids and languages worked with
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']

language_counter = Counter()

for response in lang_responses:
    language_counter.update(response.split(';'))

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

# reversing them cause we want popularity of languages in descending order
languages.reverse()
popularity.reverse()

# barh() will actually create horizontal bar-chart
plt.barh(languages, popularity)

plt.title("Most Popular Languages")
# plt.ylabel("Programming Languages")
plt.xlabel("Number of People Who Use")
plt.tight_layout()
plt.show()
```

<br><br>
### PIE CHART

```python
plt.style.use("fivethirtyeight")
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.3, 0]

# pie() will create pie chart for given data
'''
    pie(
        dataset -> on which we wanna work,
        labels(list) -> labels for each pie,
        colors(list) -> colors for each pie,
        explode(list) -> explode will explode(take out) specified pie(s) upto certain extent,
        shadow(boolean value) -> will give shadow for pie chart,
        startangle(int) -> will rotate chart upto specified angle,
        autopct(some bullshit) -> will show percentage of each pie in chart
        wedgeprops(have to pass dictionary of what prop you wanna change),
        textprops(have to pass dictionary of what prop you wanna change),
    )
'''
plt.pie(
        slices,
        labels=labels,
        explode=explode,
        shadow=True,
        startangle=90,
        autopct="%1.1f%%",
        wedgeprops={'edgecolor':'black'},
        # textprops={'color':'white'}
    )

plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```
